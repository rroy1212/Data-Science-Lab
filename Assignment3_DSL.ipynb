```python
from urllib.request import urlretrieve as download
import requests
from bs4 import BeautifulSoup
import time

link = "http://proceedings.mlr.press/v70/"
response = requests.get(link)

links = []

soup = BeautifulSoup(response.text, "html.parser")

for i in soup.findAll('div', attrs={'class':"paper"}):
    for j in i.findAll('p', attrs={'class':"links"}):
        for k in j.findAll('a'):
            if ".html" not in str(k['href']):
                          links.append(k['href'])
```

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.layout import LAParams
from pdfminer.converter import TextConverter
from pdfminer.pdfpage import PDFPage
import io
import urllib.request
import requests


def pdf_to_text(pdf_file):
    text_memory_file = io.StringIO()

    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, text_memory_file, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    # get first 3 pages of the pdf file
    for page in PDFPage.get_pages(pdf_file, pagenos=(0, 1, 2)):
        interpreter.process_page(page)
    text = text_memory_file.getvalue()
    text_memory_file.close()
    return text
```

```python
words = []

for link in links:
    response = requests.get(links[0])
    pdf_memory_file = io.BytesIO()
    pdf_memory_file.write(response.content)
    
    words.append(pdf_to_text(pdf_memory_file))
#     print(len(words))
```
